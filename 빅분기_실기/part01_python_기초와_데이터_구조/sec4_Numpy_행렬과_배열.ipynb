```python
# 두 개의 2x3 행렬 생성

mat1 = np.arange(1, 7).reshape(2,3)
mat2 = np.arange(7, 13).reshape(2, 3)
print('행렬 mat1\n', mat1)
print('행렬 mat2\n', mat2)
```

```python
my_array = np.array([mat1, mat2])
print('3차원 배열:', my_array)
print('3차원 배열 크기', my_array.shape)
# 2,2,3 : 2by3 행렬 (두 번째, 세 번째 숫자와 2와 3)이 2장 (첫번째 숫자 2) 겹쳐 있다는 듯
```

```python
# 두 개 행렬 2x3 생성
mat1 = np.arange(1, 7).reshape(2,3)
mat2 = np.arange(7, 13).reshape(2,3)

# 3차원 배열로 합치기
my_array = np.array([mat1, mat2])
print(my_array)
```

```python
# 첫 번째 2차원 배열 선택
first_slice = my_array[0, :, :]
print('첫 번째 2차원 배열: \n', first_slice)
```

```python
# 두 번째 차원의 세 번째 요소를 제외한 배열 선택
filtered_array = my_array[:, :, :-1]
print(filtered_array)
```

### transpose() 메서드
- 행렬의 전치를 실행

```python
# 원본 배열
print('원본: \n', my_array)
print('shape: ',my_array.shape)
```

```python
# 차원 변경
transposed_array = my_array.transpose(0, 2, 1) # 해당 차원 순서 (0, 1, 2)를 (0, 2, 1)로 변경을 의미 즉 두 번째 차원과 세번째 차원의 위치를 바꿔줌
print('차원 변경 배열: \n', transposed_array)
print('바뀐 차원 shape:\n', transposed_array.shape)
```

### Numpy  배열의 기본 함수

```python
# sum() : 배열의 합계 반환 axis = 0 (열) / axis = 1 (행)

import numpy as np
a = np.array([[1,2,3], [4,5,6]])

print('전체 합:', a.sum())
print('열별 합:', a.sum(axis=0))
print('행열 합:', a.sum(axis=1))
```

```python
# mean(): 원소 들의 평균

print('전체:', a.mean())
print('열별:', a.mean(axis=0))
print('행별:', a.mean(axis=1))
```

```python
a
```

```python
# max(): 배열 원소드르이 최댓값

print(a.max())
print(a.max(axis=0)) #열
print(a.max(axis=1)) #행
```

```python
# min() : 배열 원소들의 최솟값

print(a.min())
print(a.min(axis=0))
print(a.min(axis=1))
```

```python
# std(): 배열 원소들의 표준편차 반환 / ddof옵션을 사용하여 자유도 조정 ex) ddof=1의 경우 n-1로 나눠준 것을 표현하므로 표본 표준편차 의미
print('표본 표준편차:', a.std(ddof=1))
```

```python
# var(): 배열 원소들의 분산을 반환하며, ddof 옵션을 사용하여 자유도를 조정할 수 있음
print('표본 분산:', a.var(ddof=1))
```

```python
# cumsum(): 배열 원소들의 누적 합계 반환

print(a.cumsum())
print(a.cumsum(axis=0)) # 열
print(a.cumsum(axis=1)) # 행
```

```python
# cumprod: 배열 원소들의 누적 곱

print(a.cumprod())
print(a.cumprod(axis=0))
print(a.cumprod(axis=1))
```

```python
# argmax(): 최댓값의 '인덱스'를 반환한다

print('최댓값 (전체)', a.argmax())
print('최댓값 (열별)', a.argmax(axis=0))
print('최댓값 (행별)', a.argmax(axis=1))
```

```python
# argmin(): 최솟값의 인덱스 반환

print('최솟값 (전체)', a.argmin())
print('최솟값 (열별)', a.argmin(axis=0))
print('최솟값 (행별)', a.argmin(axis=1))
```

```python
# reshape(): 배열의 형상을 변경

b = np.array([1,2,3,4,5,6])
print('원본:', b)
print('변경:\n', b.reshape(2,3))
```

```python
# transpose(): 배열 전치 디폴트: 행과 열 위치 바꾸기

c = np.array([[1,2,3], [4,5,6]])
print('원본:\n', c)
print('변경:\n', c.transpose())
```

```python
# flatten() : 1차원 배열로 변경

print('1차원 배열', c.flatten())
```

```python
# clip(): 배열의 각 원소들을 주어진 최솟값과 최댓값의 범위로 자르는 역할
d = np.array([1,2,3,4,5])
print('클립된 배열', d.clip(2,4)) # d.clip(2,4): d의 각 원소를 최솟값 2와 최댓값 4로 제한
```

```python
# tolist(): 배열을 리스트로 변환
print('리스트', d.tolist())
```

```python
# astype(): 배열 원소들의 타입을 변환
e = np.array([1.1, 2.2, 3.3])
print('정수형 배열:', e.astype(int))
```

```python
# copy(): 배열의 복사본을 반환
f = d.copy()
print('복사본:', f)
```

```python
# 얕은 복사
d = np.array([1,2,3,4,5])
f = d
f[0]= 10
print('d:', d)
print('f:', f)
# d와 f가 연결되어 있어 f의 값만 바꿔도 연결되어 있는 d의 값도 바뀌는 것을 확인할 수 있음
```

```python
# 깊은 복사
d = np.array([1,2,3,4,5])
f = d.copy()
f[0] = 10
print('d:', d)
print('f:', f)

# 독립적인 변수로 된 것을 확인할 수 있음 즉 copy는 깊은 복사
```

```python
# sort()
g = np.array([3,1,2])
g.sort()
print(g)
```

```python
# argsort(): 배열의 원소들을 정렬했을 때의 인덱스 반환

h = np.array([3,1,2])
print('정렬된 인덱스:', h.argsort())
```

```python
# argsort() #2

h = np.array([10, 5, 8, 1, 7])

# 배열을 정리했을 때의 인덱스
sorted_indices = h.argsort()
print('정렬 후 인덱스:', sorted_indices)

# 정렬된 배열을 인덱스를 사용해 출력
sorted_h = h[sorted_indices]
print('정렬 후 배열', sorted_h)
```
