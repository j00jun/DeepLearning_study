# Hypothesis and Cost
### Hypothesis 
$$ H(x) = Wx + b $$

```python
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

# 초기값 지정
W = tf.Variable(2.9)
b = tf.Variable(0.5)

hypothesis = W * x_data + b
```

### cost
$$ cost(W)=\frac { 1 }{ m } \sum _{i=1}^{m}{ { (W{ x }^{ i }-y^{ i } })^{ 2 } }  $$

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_data)) # 평균(제곱(에러))
```

```python
# 전체 코드

# data
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

# W, b initialize
W = tf.Variable(2.9)
b = tf.Variable(0.5)

learning_rate = 0.01

for i in range(100+1):
    # gradient descent
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    W_grad, b_grad = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
    if i % 10 == 0:
        print("{:5}|{:10.4f}|{:10.4f}|{:10.6f}".format(i, W.numpy(), b.numpy(), cost)) # W는 1에 수렴, b는 0에 수렴할 수록 cost가 작아짐
```
