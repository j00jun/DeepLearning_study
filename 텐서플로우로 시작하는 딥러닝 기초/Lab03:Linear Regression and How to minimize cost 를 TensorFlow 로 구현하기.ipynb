```python
# cost function 구현

import numpy as np

X = np.array([1, 2, 3])
Y = np.array([1, 2, 3])

def cost_func(W, X, Y):
    c = 0
    for i in range(len(X)):
        c += (W * X[i] - Y[i]) ** 2
    return c / len(X)

for feed_W in np.linspace(-3, 5, num = 15): # linspace로 -3부터 5까지 15개의 구간을 가지게 됨
    curr_cost = cost_func(feed_W, X, Y)
    print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
```

```python
# Tensorflow로 cost funciton 구현

import tensorflow as tf

X = np.array([1, 2, 3])
Y = np.array([1, 2, 3])

def cost_func(W, X, Y):
    hypothesis = X * W
    return tf.reduce_mean(tf.square(hypothesis - Y))

W_values = np.linspace(-3, 5, num = 15)
cost_values = []

for feed_W in W_values:
    curr_cost = cost_func(feed_W, X, Y)
    cost_values.append(curr_cost)
    print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
```

```python
# 그래프로 표현
import matplotlib.pyplot as plt

plt.plot(W_values, cost_values, marker='o')
plt.title('Cost function graph')
plt.xlabel('W')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```
